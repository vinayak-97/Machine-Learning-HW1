```python
df = df.rename(columns={"male":"Sex"})
df
```

```python
df = df.rename(columns={"Q":"Em_Q","S":"Em_S"})
df
```

```python
from sklearn.preprocessing import MinMaxScaler

age_scaler = MinMaxScaler()
age_scaler.fit(df[["Age"]])
df["Age"] = age_scaler.transform(df[["Age"]])
df
```

```python
from sklearn.preprocessing import MinMaxScaler

fare_scaler = MinMaxScaler()
fare_scaler.fit(df[["Fare"]])
df["Fare"] = fare_scaler.transform(df[["Fare"]])
df
```

```python
X = df.drop("Survived",axis=1)
y = df["Survived"]
```

```python
from sklearn.tree import DecisionTreeClassifier

Dt_model = DecisionTreeClassifier()
Dt_model.fit(X,y)
```

```python
dt_score = round(Dt_model.score(X,y)*100,2)
print("Decision Tree score: ",dt_score)
```

```python
from sklearn.ensemble import RandomForestClassifier

Rf_model = RandomForestClassifier()
Rf_model.fit(X,y)
Rf_score = round(Rf_model.score(X,y)*100,2)
print("Random Forest Score: ",Rf_score)
```

```python
from sklearn.neighbors import KNeighborsClassifier

Knn_model = KNeighborsClassifier()

Knn_model.fit(X,y)
Knn_score = round(Knn_model.score(X,y)*100,2)
print("KNN Classifier score: ",Knn_score)
```

```python
from sklearn.linear_model import LogisticRegression

Lr_model = LogisticRegression()

Lr_model.fit(X,y)
Lr_score = round(Lr_model.score(X,y)*100,2)
print("LogisticRegression score: ",Lr_score)
```

```python
from sklearn.svm import LinearSVC

Ls_model = LinearSVC()
Ls_model.fit(X,y)
Ls_score = round(Ls_model.score(X,y)*100,2)
print("LinearSVC Score: ",Ls_score)
```

```python
from sklearn.linear_model import Perceptron

P_model = Perceptron(penalty="l2",alpha = 0.00001,eta0 = 2)
P_model.fit(X,y)
P_score = round(P_model.score(X,y)*100,2)
print("Perceptron model Score: ",P_score)
```

```python
from sklearn.svm import SVC

Svc_model = SVC()
Svc_model.fit(X,y)
Svc_score=round(Svc_model.score(X,y)*100,2)
print("Support Vector Machine model score: ",Svc_score)
```

```python
from sklearn.linear_model import SGDClassifier

Sgdc_model = SGDClassifier()
Sgdc_model.fit(X,y)
Sgdc_score = round(Sgdc_model.score(X,y)*100,2)
print("Stochastic Gradient Decent model score: ",Sgdc_score)
```

```python
from sklearn.naive_bayes import GaussianNB

Gnb_model = GaussianNB()
Gnb_model.fit(X,y)
Gnb_score = round(Gnb_model.score(X,y)*100,2)
print("Naive Bayes model score: ",Gnb_score)
```

```python
models = pd.DataFrame({
    'Model': ['Support Vector Machines', 'KNN', 'Logistic Regression', 
              'Random Forest', 'Naive Bayes', 'Perceptron', 
              'Stochastic Gradient Decent', 'Linear SVC', 
              'Decision Tree'],
    'Score':[Svc_score,Knn_score,Lr_score,Rf_score,Gnb_score,P_score,Sgdc_score,Ls_score,dt_score]})

models.sort_values(by='Score', ascending=False)
```
